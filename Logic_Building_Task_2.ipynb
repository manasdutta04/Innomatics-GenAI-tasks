## Problem 1: Unique Words in a Sentence
### Problem Statement
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves


```python
# Get input sentence from the user
sentence = input("Enter sentence: ")

# Split the sentence into words and convert to lowercase
words = sentence.lower().split()

# Convert the list of words into a set to remove duplicates
unique_words = set(words)

# Display the count of unique words
print(f"Unique words count: {len(unique_words)}")

# Display the unique words themselves
print(f"Unique words: {unique_words}")
```

## Problem 2: Highest Salary from Employee Data
### Problem Statement
Store employee names and their salaries, and determine which employee has the highest salary.


```python
# Example Input Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000,
    "John": 80000,
    "Maria": 70000
}

# Initialize variables to keep track of the highest salary and the employee name
highest_salary = 0
highest_earner = ""

# Iterate through the dictionary to find the highest salary
for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_earner = name

# Display the employee with the highest salary
print(f"Highest Salary: {highest_earner} - {highest_salary}")
```

## Problem 3: Find Maximum and Minimum Values
### Problem Statement
Find the maximum and minimum values from a list of numbers.


```python
# Example Input List of Numbers
numbers = [45, 22, 89, 10, 66, 5, 100, 33]

# Initialize max_val and min_val with the first element of the list
if not numbers:
    print("The list is empty.")
else:
    max_val = numbers[0]
    min_val = numbers[0]

    # Iterate through the list to find the maximum and minimum values
    for num in numbers:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num

    # Display the list, maximum, and minimum values
    print(f"List: {numbers}")
    print(f"Max: {max_val}")
    print(f"Min: {min_val}")
```

## Problem 4: Count Products Above a Price Threshold
### Problem Statement
Count how many products have a price greater than 1000.


```python
# Example Input List of Prices
prices = [450, 1200, 899, 1500, 300, 1001, 999]

# Initialize a counter for products above the threshold
count_above_threshold = 0
threshold = 1000

# Iterate through the prices and count those above the threshold
for price in prices:
    if price > threshold:
        count_above_threshold += 1

# Display the count
print(f"Products above {threshold}: {count_above_threshold}")
```

## Problem 5: Calculate Attendance Percentage
### Problem Statement
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
# Example Input Attendance List
attendance = ["P", "P", "A", "P", "P", "A", "P"]

# Calculate total days and present days
total_days = len(attendance)
present_days = attendance.count("P")

# Calculate attendance percentage
if total_days > 0:
    attendance_percentage = (present_days / total_days) * 100
    print(f"Attendance Percentage: {attendance_percentage:.1f}")
else:
    print("Attendance list is empty, cannot calculate percentage.")
```

## Problem 7: Count Character Frequency
### Problem Statement
Count how many times each character appears in a given string.


```python
# Example Input String
input_string = "python"

# Initialize an empty dictionary to store character counts
char_counts = {}

# Iterate through the string and count character frequencies
for char in input_string:
    char_counts[char] = char_counts.get(char, 0) + 1

# Display the character frequencies
print(f"Character Frequencies: {char_counts}")
```

## Problem 10: Calculate Average Marks
### Problem Statement
Calculate the average marks of a student from a list of marks.


```python
# Example Input List of Marks
marks = [85, 90, 70, 65, 80, 55, 95]

# Calculate the sum of marks
total_marks = sum(marks)

# Calculate the number of marks
num_marks = len(marks)

# Calculate the average marks
if num_marks > 0:
    average_marks = total_marks / num_marks
    print(f"Average Marks: {average_marks:.1f}")
else:
    print("The list of marks is empty, cannot calculate average.")
```
