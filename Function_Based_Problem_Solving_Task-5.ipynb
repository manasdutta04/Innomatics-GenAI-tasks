## Problem 1: Social Media – Post Engagement Analyzer
Problem Statement
Create a function to analyze engagement on a social media post.


```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_post_engagement(total_likes):
    """
    Analyzes social media post engagement based on total likes.

    Args:
        total_likes (int): The total number of likes on a post.

    Returns:
        tuple: A tuple containing total likes and the post status.
    """
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    return total_likes, status

# Example Usage for Problem 1:
likes, post_status = analyze_post_engagement(1050)
print(f"Total Likes: {likes}")
print(f"Post Status: {post_status}")

likes, post_status = analyze_post_engagement(500)
print(f"\nTotal Likes: {likes}")
print(f"Post Status: {post_status}")
```

## Problem 3: Agriculture – Rainfall Adequacy Checker
Problem Statement
Create a function that analyzes seasonal rainfall data.


```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
def check_rainfall_adequacy(rainfall_data, required_level):
    """
    Analyzes seasonal rainfall data to determine adequacy.

    Args:
        rainfall_data (list): A list of rainfall measurements.
        required_level (int): The required average rainfall level.

    Returns:
        tuple: A tuple containing the average rainfall and its status.
    """
    if not rainfall_data:
        return 0, "No Rainfall Data"

    average_rainfall = sum(rainfall_data) / len(rainfall_data)

    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    return int(average_rainfall), status # Cast to int for output format

# Example Usage for Problem 3:
rainfall_measurements = [65, 70, 80, 75, 70]
required_rainfall = 72
avg_rain, rainfall_status = check_rainfall_adequacy(rainfall_measurements, required_rainfall)
print(f"Average Rainfall: {avg_rain}")
print(f"Rainfall Status: {rainfall_status}")

rainfall_measurements_low = [50, 60, 55, 65]
required_rainfall_low = 70
avg_rain_low, rainfall_status_low = check_rainfall_adequacy(rainfall_measurements_low, required_rainfall_low)
print(f"\nAverage Rainfall: {avg_rain_low}")
print(f"Rainfall Status: {rainfall_status_low}")
```

## Problem 4: Social Media – Duplicate Account Detection
Problem Statement
Create a function to detect duplicate usernames in a system.


```python
# Problem 4: Social Media – Duplicate Account Detection
def detect_duplicate_accounts(usernames):
    """
    Detects duplicate usernames in a system.

    Args:
        usernames (list): A list of usernames.

    Returns:
        str: "Yes" if duplicates exist, "No" otherwise.
    """
    if len(usernames) != len(set(usernames)):
        return "Yes"
    else:
        return "No"

# Example Usage for Problem 4:
users1 = ["user1", "user2", "user3", "user1"]
duplicate_status1 = detect_duplicate_accounts(users1)
print(f"Duplicate Accounts Found: {duplicate_status1}")

users2 = ["userA", "userB", "userC"]
duplicate_status2 = detect_duplicate_accounts(users2)
print(f"Duplicate Accounts Found: {duplicate_status2}")
```

## Problem 6: Agriculture – Premium Crop Price Filter
Problem Statement
Create a function that filters crop prices above a premium threshold.


```python
# Problem 6: Agriculture – Premium Crop Price Filter
def filter_premium_crops(prices):
    """
    Filters crop prices to identify premium crops based on a threshold.

    Args:
        prices (list): A list of crop prices.

    Returns:
        list: A list of prices considered premium crops.
    """
    premium_crops = [price for price in prices if price > 2000]
    return premium_crops

# Example Usage for Problem 6:
crop_prices1 = [1800, 2500, 1500, 3200, 2100]
premium_list1 = filter_premium_crops(crop_prices1)
print(f"Premium Crops: {premium_list1}")

crop_prices2 = [1000, 1200, 1900]
premium_list2 = filter_premium_crops(crop_prices2)
print(f"Premium Crops: {premium_list2}")
```

## Problem 7: System Monitoring – Application Health Checker
Problem Statement
Create a function to evaluate application health based on error count.


```python
# Problem 7: System Monitoring – Application Health Checker
def check_application_health(error_count):
    """
    Evaluates application health based on the number of errors.

    Args:
        error_count (int): The number of errors detected in the application.

    Returns:
        tuple: A tuple containing the error count and the system status.
    """
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    return error_count, status

# Example Usage for Problem 7:
errors, system_status = check_application_health(7)
print(f"Error Count: {errors}")
print(f"System Status: {system_status}")

errors_healthy, system_status_healthy = check_application_health(0)
print(f"\nError Count: {errors_healthy}")
print(f"System Status: {system_status_healthy}")

errors_minor, system_status_minor = check_application_health(3)
print(f"\nError Count: {errors_minor}")
print(f"System Status: {system_status_minor}")
```

## Problem 8: Banking – Daily Transaction Limit Checker
Problem Statement
Create a function to validate daily transaction amount.


```python
# Problem 8: Banking – Daily Transaction Limit Checker
def check_transaction_limit(transaction_amount):
    """
    Validates a daily transaction amount against a predefined limit.

    Args:
        transaction_amount (int): The amount of the transaction.

    Returns:
        tuple: A tuple containing the transaction amount and its status.
    """
    DAILY_LIMIT = 50000 # Assuming the limit is ₹50,000

    if transaction_amount <= DAILY_LIMIT:
        status = "Approved"
    else:
        status = "Rejected"
    return transaction_amount, status

# Example Usage for Problem 8:
transaction_amt, transaction_status = check_transaction_limit(60000)
print(f"Transaction Amount: {transaction_amt}")
print(f"Transaction Status: {transaction_status}")

transaction_amt_approved, transaction_status_approved = check_transaction_limit(45000)
print(f"\nTransaction Amount: {transaction_amt_approved}")
print(f"Transaction Status: {transaction_status_approved}")
```

## Problem 9: E-Learning – Student Attendance Eligibility System
Problem Statement
Create a function to calculate attendance percentage and check exam eligibility.


```python
# Problem 9: E-Learning – Student Attendance Eligibility System
def check_attendance_eligibility(attended_classes, total_classes):
    """
    Calculates attendance percentage and determines exam eligibility.

    Args:
        attended_classes (int): The number of classes attended by the student.
        total_classes (int): The total number of classes conducted.

    Returns:
        tuple: A tuple containing the attendance percentage and eligibility status.
    """
    if total_classes == 0:
        return 0.0, "No Classes Conducted"

    attendance_percentage = (attended_classes / total_classes) * 100

    if attendance_percentage >= 75:
        eligibility_status = "Eligible"
    else:
        eligibility_status = "Not Eligible"

    return round(attendance_percentage, 1), eligibility_status

# Example Usage for Problem 9:
attendance_pct, eligibility = check_attendance_eligibility(75, 100)
print(f"Attendance Percentage: {attendance_pct}")
print(f"Exam Eligibility: {eligibility}")

attendance_pct_not_eligible, eligibility_not_eligible = check_attendance_eligibility(60, 100)
print(f"\nAttendance Percentage: {attendance_pct_not_eligible}")
print(f"Exam Eligibility: {eligibility_not_eligible}")

attendance_pct_borderline, eligibility_borderline = check_attendance_eligibility(74, 100)
print(f"\nAttendance Percentage: {attendance_pct_borderline}")
print(f"Exam Eligibility: {eligibility_borderline}")
```

## Problem 10: Smart Electricity Bill Analyzer
Problem Statement
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.


```python
# Problem 10: Smart Electricity Bill Analyzer
def calculate_electricity_bill(units_consumed):
    """
    Calculates the electricity bill and classifies usage level.

    Args:
        units_consumed (int): The number of electricity units consumed.

    Returns:
        tuple: A tuple containing the total bill amount and usage classification.
    """
    total_bill = 0
    usage_classification = ""

    if units_consumed <= 100:
        total_bill = units_consumed * 3
    elif units_consumed <= 200:
        total_bill = (100 * 3) + ((units_consumed - 100) * 5)
    else:
        total_bill = (100 * 3) + (100 * 5) + ((units_consumed - 200) * 7)

    if total_bill < 500:
        usage_classification = "Low Usage"
    elif 500 <= total_bill <= 1500:
        usage_classification = "Moderate Usage"
    else:
        usage_classification = "High Usage"

    return total_bill, usage_classification

# Example Usage for Problem 10:
bill, usage = calculate_electricity_bill(50)
print(f"Total Bill: ₹{bill}")
print(f"Usage Level: {usage}")

bill, usage = calculate_electricity_bill(150)
print(f"\nTotal Bill: ₹{bill}")
print(f"Usage Level: {usage}")

bill, usage = calculate_electricity_bill(250)
print(f"\nTotal Bill: ₹{bill}")
print(f"Usage Level: {usage}")

bill, usage = calculate_electricity_bill(300)
print(f"\nTotal Bill: ₹{bill}")
print(f"Usage Level: {usage}")
```
